```python
import pandas as pd
import numpy as np
```

```python
old = pd.read_excel('Old_Judiciario.xlsx')
old['data_julgamento'] = np.nan
old['data_publicacao'] =  np.nan
old_dates = pd.read_excel('Old_Judiciario_Dates.xlsx')
```

```python
julgamento = []
publicacao = []
ids = list(set(list(old_dates['NÚMERO ID'])))

for id_number in ids:
    try:
        julgamento.append(old_dates[old_dates['NÚMERO ID'] == id_number]['DATA DO\nJULGAMENTO'].iloc[0])
        publicacao.append(old_dates[old_dates['NÚMERO ID'] == id_number]['DATA DA PUBLICAÇÃO'].iloc[0])
    except:
        pass
```

```python
for id_number in ids:
    try:
        old.loc[old[old['numero_id'] == id_number].index, 'data_julgamento']= julgamento[ids.index(id_number)]
        old.loc[old[old['numero_id'] == id_number].index, 'data_publicacao']= julgamento[ids.index(id_number)]
    except:
        pass
```

```python
old
```

```python
forma_mencao = []
for i in range(len(old)):
    if old.iloc[i]['mencao_imprecisa'] == 'Yes':
        forma_mencao.append('Imprecise mention to a WHO document')
    elif old.iloc[i]['mencao_meramente_generica'] == 'Yes':
        forma_mencao.append('Generic mention to WHO')
    elif old.iloc[i]['mencao_especifica'] == 'Yes':
        forma_mencao.append('Precise mention to a WHO document')
    else:
        forma_mencao.append('N/A')
old['forma_mencao'] = forma_mencao
```

```python
for i in range(len(old)):
    if type(old.iloc[i]['major_health_issue_other_specify']) == str:
        old.loc[i, 'major_health_issue'] = old.iloc[i]['major_health_issue'] + old.iloc[i]['major_health_issue_other_specify']
```

```python
new = pd.read_excel('New_Judiciario.xlsx')
```

```python
tribunal = list(old['tribunal']) + list(new['Órgão'])
numero_id = list(old['numero_id']) + list(new['ID (número do processo)'])
link = list(old['link']) + list(new['Link'])
tipo = list(old['tipo']) + list(new['Tipo de decisão (colegiada ou monocrática)'])
decisao = list(old['decisao']) + list(new['Tipo de decisão (unânime ou majoritária)'])
relator = list(old['relator']) + list(new['Relator'])
data_julgamento = list(old['data_julgamento']) + list(new['Data de julgamento'])
data_publicacao = list(old['data_publicacao']) + list(new['Data de publicação'])

forma_mencao = list(old['forma_mencao']) + list(new['Forma da Menção'])
local_da_mencao = list(old['mencao_onde_pt']) + list(new['Local da menção à OMS'])
nome_do_documento_citado = list(old['nome_do_documento_citado_se_houver']) + list(new['Nome do documento da OMS citado'])
soft_hard = list(old['documento_hard_ou_soft']) + list(new['Tipo de documento: hard (vinculante) ou soft (não vinculante)'])
ministro_do_voto_que_menciona = list(old['ministro_do_voto_que_menciona']) + list(new['Nome do Ministro que proferiu o voto que contém a menção à OMS'])
forma_da_mencao_direta_or_indireta = list(old['forma_da_mencao_direct_or_indirect']) + list(new['Tipo de menção (direta ou indireta)'])
trecho_citado = list(old['trecho_citado']) + list(new['Menções'])
major_health_issue = list(old['major_health_issue']) + list(new['Health Topic'])
```

```python
jus = pd.DataFrame(data = {'tribunal': tribunal, 'numero_id': numero_id, 'link': link, 'tipo': tipo,
                          'decisao': decisao, 'relator': relator, 'data_julgamento': data_julgamento,
                          'data_publicacao': data_publicacao, 'local_da_mencao': local_da_mencao,
                          'nome_do_documento_citado': nome_do_documento_citado, 'soft_hard': soft_hard,
                          'ministro_do_voto_que_menciona': ministro_do_voto_que_menciona, 
                          'forma_da_mencao_direta_or_indireta': forma_da_mencao_direta_or_indireta,
                          'trecho_citado': trecho_citado, 'forma_mencao': forma_mencao,
                          'major_health_issue': major_health_issue})
jus
```

```python
jus.replace('IDEM', method = 'ffill', inplace = True)
jus.replace('idem', method = 'ffill', inplace = True)
jus.replace('Idem', method = 'ffill', inplace = True)
```

```python
local = jus['local_da_mencao']

rel = []
for i, local in enumerate(local):
    if type(local) != str:
        rel.append(local)
    elif 'relator' in local.lower():
        rel.append('Voto vencedor')
    else:
        rel.append(local)
jus['local_da_mencao'] = rel
```

```python
jus.to_excel('Judiciario.xlsx', index = False)
```
