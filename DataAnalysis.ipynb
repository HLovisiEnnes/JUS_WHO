```python
covid = []
for ht in list(jus['major_health_issue']):
    if type(ht) == str:
        if 'covid' in ht.lower():
            covid.append(1)
        else:
            covid.append(0)
    else:
        covid.append(0)
```

```python
jus['COVID'] = covid
```

```python
month = []
year = []
for date in list(jus['data_julgamento']):
    if type(date) == datetime.datetime:
        month.append(date.strftime('%m/%Y'))
        year.append(date.strftime('%Y'))
    elif type(date) == str:
        month.append(date.split('/')[1] + '/' + date.split('/')[-1])
        year.append(date.split('/')[-1])
    else:
        month.append(date)
        year.append(date)
```

```python
jus['year'] = year
jus_covid = jus[jus['COVID'] == 1]
jus_n_covid = jus[jus['COVID'] == 0]
```

```python
years = [str(i) for i in range(1988, 2023)]
years_jus = []
years_covid = []
years_n_covid = []
for year in years:
    years_jus.append(list(jus['year']).count(year))
    years_covid.append(list(jus_covid['year']).count(year))
    years_n_covid.append(list(jus_n_covid['year']).count(year))
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(20, 5)
ax.plot(years, years_covid, '-o', label = 'Mentions COVID-19')
ax.plot(years, years_n_covid, '-o', label = 'Does not mention COVID-19')
ax.bar(years, years_jus, color = 'gray', alpha = 0.5,
      label = 'All mentions')
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years, rotation='vertical', fontsize= 10)
ax.set_ylabel('Yearly number of citations to WHO',  fontsize= 15)
ax.set_title('Citations to WHO',  fontsize= 15)
plt.legend()
plt.savefig('graphs/dates.png')
plt.show()
```
